```python
import pandas
import requests
import json
import csv


indicators_who = [
    "Age-standardized suicide rates (per 100 000 population)",
    "Primary data availability for UHC Service Coverage Index (SDG 3.8.1) (%)",
    "Mean BMI (kg/m&#xb2;) (crude estimate)",
    "Prevalence of anaemia in women of reproductive age (aged 15-49) (%)",
    "Poliomyelitis - number of reported cases",
    "Estimates of rates of homicides per 100 000 population",
    "Prevalence of thinness among children and adolescents, BMI < -2 standard deviations below the median (crude estimate) (%)",
    "Estimate of current tobacco use prevalence (%) (age-standardized rate)",
    "Prevalence of obesity among adults, BMI &GreaterEqual; 30 (age-standardized estimate) (%)"
]
dict_who = {}

indicators_wb = [
    "EG.ELC.ACCS.ZS",
    "NY.GDP.PCAP.CD",
    "SP.RUR.TOTL.ZS",
    "SH.DYN.MORT",
    "EN.ATM.PM25.MC.M3",
    "SP.POP.TOTL"
]


request = requests.get(url="https://ghoapi.azureedge.net/api/Indicator")
request.raise_for_status()
print(request)
list_api = request.json()
```

```python
for i in list_api["value"]:
    if i["IndicatorName"] in indicators_who:
        dict_who[i["IndicatorCode"]] = i["IndicatorName"]

for i in dict_who:
    print(i, dict_who[i])
```
